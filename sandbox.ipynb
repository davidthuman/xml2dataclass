```python
def to_class_name(name: str) -> str:
    
    for char in ["-", ".", ":"]:
        name = name.replace(char, "_")
        
    words = name.split("_")
    
    return "".join(map(lambda word: word[0].upper() + word[1:], words))
```

```python
@dataclass
class FieldLite:
    
    name: str
    type_hint: str
    
    def to_py_str(self):
        return f"{self.name}: {self.type_hint}"

@dataclass
class DataClass:
    
    name: str
    attributes: list[FieldLite]
    fields: list[FieldLite]
    
    def to_py_str(self):
        
        fields_py_str = "\t"
    
        for field in [*self.attributes, *self.fields]:
            
            fields_py_str += field.to_py_str() + "\n\t"
        
        py_str = f"""\
@dataclass
class {to_class_name(self.name)}:

{fields_py_str}
"""
        return py_str
```

```python
print("dataclass Name: ", root.tag)
print("dataclass Attributes: ", root.attrib)
print("dataclass Fields: ", [child.tag for child in root])
```

```python

element = root.find("country")

def parse_element(element: ET.Element):

    # parse children elements

    child_count = {}

    for child in element:
        
        if child.tag not in child_count:
            child_count.update({child.tag: 1})
        else:
            child_count.update({child.tag: child_count.get(child.tag) + 1})
            
            
    # Determine type of child element
    # - if no children element -> basic type
    # - if has attributes -> needs dataclass
    # - if has children element -> needs dataclass

    child_needs_dataclass = {}

    for key in child_count.keys():
        
        # Check all instances of child element, in case the parent element has different structured child elements
        
        for child in element.findall(key):
            
            has_attributes = True if len(child.attrib) > 0 else False
            has_children = True if len(child) > 0 else False
            
            needs_dataclass = has_attributes or has_children
            
            if key not in child_needs_dataclass:
                child_needs_dataclass.update({key: needs_dataclass})
            else:
                child_needs_dataclass.update({key: child_needs_dataclass.get(key) or needs_dataclass})
                
            
    # Create element's       
            
            
    # if child element has more than 1 instance, field on parent element will be a list of the child element type
    
    attributes = []
    fields = []

    for attribute_name, attribute_value in element.attrib.items():
        
        attributes.append(FieldLite(attribute_name, "str"))
    
    for tag, needs_dataclass in child_needs_dataclass.items():
                
        # Field Name
        if child_count.get(tag) == 1:
            field_name = tag
        else:
            field_name = tag + "s"
         
        # Field Type
        if not needs_dataclass and child_count.get(tag) == 1:
            field_type = "str"
        elif not needs_dataclass and child_count.get(tag) > 1:
            field_type = "list[str]"
        elif needs_dataclass and child_count.get(tag) == 1:
            tag_type = parse_element(element=element.find(tag)).name
            field_type = f"{tag_type}"
        elif needs_dataclass and child_count.get(tag) > 1:
            tag_type = parse_element(element=element.find(tag)).name
            field_type = f"list[{tag_type}]"
                    
        fields.append(FieldLite(field_name, field_type))
        
        
    data_class = DataClass(to_class_name(element.tag), attributes, fields)
    
    print(data_class.to_py_str())
    
    return data_class
    
parse_element(root)
```
